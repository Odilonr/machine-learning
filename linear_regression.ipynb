```python
from nba_api.stats import endpoints
```

```python
import seaborn as sns; sns.set_style('whitegrid');
```

```python
seasons = np.arange(2000, 2020)
seasons2 = np.arange(0, 21)
```

```python
data = endpoints.leagueleaders.LeagueLeaders(season = '1999-00').league_leaders.get_data_frame()

data['season_start'] = 1999
data['season_end'] = 2000

for i, v in enumerate(seasons):
    temp = endpoints.leagueleaders.LeagueLeaders(season = f'{v}-{str(i+1).zfill(2)}').league_leaders.get_data_frame()
    temp['season_start'] = v
    temp['season_end'] = v + 1
    data = data.append(temp)

df = data

df
```

```python
print(df['season_start'].unique())
print(df['season_end'].unique())
```

```python
df = data.loc[:,['season_start','PLAYER','PLAYER_ID','GP','MIN','FGM','FGA','FG3M','FG3A','FTM','FTA','REB','AST','STL','BLK','TOV','PF','PTS','EFF','AST_TOV','STL_TOV']].reset_index()

df['PTS_NextYear'] = df.sort_values(by = ['PLAYER_ID', 'season_start']).groupby('PLAYER_ID')['PTS'].shift(-1)

df = df.loc[df['MIN'] > 1600]

df = df.loc[df['PTS_NextYear'].notnull()]

df.loc[df['PLAYER'] == 'LeBron James', ['PLAYER','season_start','PTS','PTS_NextYear','MIN','FGA']].sort_values(by='season_start')
```

```python
df.corr()['PTS_NextYear'].sort_values(ascending = False)
```

```python
from sklearn.model_selection import train_test_split

X = df[['PTS', 'FGM', 'FGA', 'FTM', 'FTA', 'EFF', 'TOV', 'MIN']].values
y = df[['PTS_NextYear']].values

print(f'Original Data Shape - X: {X.shape}, Y: { y.shape}')

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.2)

print(f'Train Data Shape - X: {X_train.shape}, Y: {y_train.shape}')
print(f'Test Data Shape - X: {X_test.shape}, Y: {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error


lr = LinearRegression()


lr.fit(X_train, y_train)


```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_train)

print('Mean number of Points:', y_train[:,0].mean())
print('Mean absolute error', mean_absolute_error(y_pred, y_train))
```

```python
y_pred1 = lr.predict(X_test)

print('Mean of Points: ', y_test[:,0].mean())
print('Mean abslolute error:', mean_absolute_error(y_pred1, y_test))

```
